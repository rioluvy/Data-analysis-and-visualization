## 문자열 다루기 실전

```python
df = pd.read_csv('../data/matzipData.csv', encoding='utf-8')
df
```

```python
df.info()
```

```python
df_food = df['Food'].astype(str).str.split('/', expand=True)
df_food
```

```python
df_food.columns = ['메뉴1', '메뉴2', '메뉴3']
```

```python
df = pd.concat((df, df_food), axis=1)
df
```

```python
df['Parking'].unique()
```

```python
# 딱히 필요없는 정보들은 빈 문자열로 변환
df_parking = df['Parking'].str.replace('주차권','')\
                        .str.replace('제공','')\
                        .str.replace('가능', '')
df_parking.unique()
```

```python
# '시간 무료' -> '시간'
# '당일 무료' -> '당일'
df_parking = df_parking.str.replace('시간 무료','시간')\
                        .str.replace('당일 무료','당일')
df_parking.unique()
```

```python
df_parking = df_parking.str.split(expand=True)
```

```python
df_parking.shape
```

```python
df_parking.iloc[:,0].value_counts()
```

```python
df_parking.iloc[:,1].value_counts()
```

```python
df_parking.columns = ['주차형태','무료시간']
```

```python
df_parking.head()
```

```python
df_parking['주차형태'].value_counts()
```

```python
df_parking['무료시간'].value_counts()
```

```python
df = pd.concat((df, df_parking), axis=1)
df.head()
```
