## Using colormap when creating colorbar

```python
np.random.seed(1)
data = np.random.rand(14,16)
```

```python
fig, ax = plt.subplots()
im = ax.imshow(data, cmap='viridis')
ax.set_xlabel("x")

cb = fig.colorbar(im, pad=0.05)
```

```python
fig, ax = plt.subplots()
im = ax.imshow(data, cmap='magma')
ax.set_xlabel("x")

cb = fig.colorbar(im, pad=0.05)
```

### Draw Discrete Colormap using named Colormap

```python
fig, ax = plt.subplots()

from matplotlib import cm

# get_cmap before create ScalarMappable
viridis_discrete = cm.get_cmap('viridis', 8)

im = ax.imshow(data, cmap=viridis_discrete)

ax.set_xlabel("x")

cb = fig.colorbar(im, pad=0.05)
```

### Custom Colormap

### 여러 색상 혼합

#### 1. 이산형

```python
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
my_discrete_cmap = ListedColormap(["red", "magenta", "purple", "orange", "dodgerblue"])

fig, ax = plt.subplots()
im = ax.imshow(data, cmap=my_discrete_cmap)

ax.set_xlabel("x")
cb = fig.colorbar(im, pad=0.05)
```

#### 2. 연속형

```python
from matplotlib.colors import ListedColormap, LinearSegmentedColormap

colors = ["red", "magenta", "purple", "orange", "dodgerblue"]
nodes = [0.0,0.25,0.5,0.75,1.0]
my_continuous_cmap = LinearSegmentedColormap.from_list("mycmap", list(zip(nodes, colors)))

fig, ax = plt.subplots()
im = ax.imshow(data, cmap=my_continuous_cmap)

ax.set_xlabel("x")
cb = fig.colorbar(im, pad=0.05)
```

### 두 개 색상 혼합

```python
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
colors = ["white", "dodgerblue"]
nodes = [0.0,1.0]
twocolor_cmap = LinearSegmentedColormap.from_list("mycmap", list(zip(nodes, colors)))

fig, ax = plt.subplots()
im = ax.imshow(data, cmap=twocolor_cmap)

ax.set_xlabel("x")
cb = fig.colorbar(im, pad=0.05)
```

```python
colors = ["magenta", "orange"]
nodes = [0.0,1.0]
twocolor_cmap = LinearSegmentedColormap.from_list("mycmap", list(zip(nodes, colors)))

fig, ax = plt.subplots()
im = ax.imshow(data, cmap=twocolor_cmap)

ax.set_xlabel("x")
cb = fig.colorbar(im, pad=0.05)
```
