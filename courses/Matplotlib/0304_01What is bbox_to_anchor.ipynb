```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as patches
```

```python
x0, y0, width, height = 0.2, 0.2, 0.6, 0.3
```

```python
x = np.arange(0.1, 4, 0.1)
y = 1.0/x

fig = plt.figure(figsize=(6, 4))

ax = fig.add_subplot()
ax.plot(x,y, label='1/x')
loc_str = 'lower left'
ax.legend(loc=loc_str, bbox_to_anchor=(x0,y0,width,height),edgecolor='g', fontsize='large', framealpha=0.5, borderaxespad=0)
ax.add_patch(patches.Rectangle((x0,y0), width, height, color='r',fill=False, transform=ax.transAxes))
ax.text(0.6, 0.2, s="loc = '{}'".format(loc_str),transform=ax.transAxes)
```
