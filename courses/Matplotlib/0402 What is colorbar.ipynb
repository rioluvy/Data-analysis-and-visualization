```python
import matplotlib.pyplot as plt
import numpy as np; np.random.seed(1)

data = np.random.rand(14,16)
```

```python
fig, ax = plt.subplots()
im = ax.imshow(data)
ax.set_xlabel("x")

cb = fig.colorbar(im, pad=0.05)
```

## Manipulaing a colorbar

### changing colormaps
 - "viridis", "plasma", "inferno", "magma", cividis"

```python
fig, ax = plt.subplots()
im = ax.imshow(data, cmap = "magma")
ax.set_xlabel("x")

cb = fig.colorbar(im, pad=0.05)
```

### adjusting clim : set_clim(vmin=None, vmax=None)

```python
fig, ax = plt.subplots()
im = ax.imshow(data, cmap = "magma")
im.set_clim(vmin=0.2, vmax=0.6)
ax.set_xlabel("x")

cb = fig.colorbar(im, pad=0.05)
```

### adjusting the colorbar positions

```python
fig, ax = plt.subplots()
im = ax.imshow(data, cmap = "plasma")
ax.set_xlabel("x")

cb = fig.colorbar(im, pad=0.3)
```

```python
fig, ax = plt.subplots()
im = ax.imshow(data, cmap = "plasma")
ax.set_xlabel("x")

cb = fig.colorbar(im, pad=0.25, orientation="horizontal")
```

```python
fig, ax = plt.subplots(constrained_layout=True)
im = ax.imshow(data, cmap = 'plasma')
fig.colorbar(im, ax=ax,location='left')
```
