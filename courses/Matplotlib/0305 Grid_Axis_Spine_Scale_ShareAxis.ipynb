### 3. Control Spines, https://wikidocs.net/141849

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(-3, 3, 100)
```

```python
fig, ax = plt.subplots()
ax.set_title('Mean Squared Error', pad=20)
ax.plot(x, x**2, color='#4799FF', linewidth=5)
```

```python
#ax.spines['left'].set_position('center')        # 왼쪽 축을 가운데 위치로 이동
ax.spines['left'].set_position(('axes', 0.5))
ax.spines['right'].set_visible(False)          # 오른쪽 축을 보이지 않도록
ax.spines['top'].set_visible(False)            # 위 축을 보이지 않도록
ax.spines['bottom'].set_position(('data', 0))   # 아래 축을 데이터 0의 위치로 이동

#ax.set_xlim([-3, 3])
#ax.set_ylim([0, 3])
#ax.set_xticks([-3, -2, -1, 0, 1, 2, 3])
#ax.set_yticks([1, 2, 3])

fig
```

### 4. Change Axis Scales, https://wikidocs.net/137913

- linear
- log
- symlog (= symmetrical log scale)
- logit

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = x ** 2
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_yscale('log')
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_yscale('symlog')
```

```python
x = np.linspace(0, 1, 100)
y = x ** 2

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x, y)
ax.set_yscale('logit')
#ax.set_ylim([10^(-4), 1-10^(-2)])
```

### 5. Shared Axis, https://wikidocs.net/141535

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

x = np.arange(0, 3)
y1 = x + 1
y2 = -x - 1

fig, ax1 = plt.subplots()
ax1.set_xlabel('X-Axis')
ax1.set_ylabel('1st Y-Axis')
line1, = ax1.plot(x, y1, color='green')

ax2 = ax1.twinx()
ax2.set_ylabel('2nd Y-Axis')
line2, = ax2.plot(x, y2, color='deeppink')

ax1.legend((line1, line2), ('1st Data', '2nd Data'), loc='center right', fontsize=8)
```
