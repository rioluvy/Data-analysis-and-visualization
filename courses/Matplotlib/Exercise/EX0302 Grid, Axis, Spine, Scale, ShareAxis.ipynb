# Grid, Axis, Spin, Scale, ShareAxis 에 대해 알아보자

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Grid - 그래프에 격자를 표시한다.

```python
fig =plt.figure()
ax = fig.add_subplot()
ax.grid(True)
```

```python
ax.grid(True, alpha=0.5, linestyle='--', color='r')
fig
```

## 2. Axis 그래프의 축을 말하며, 범위 지정 가능
- axis(), set_xlim(), set_ylim()

```python
fig =plt.figure()
ax = fig.add_subplot()
ax.grid(True)
ax.axis([-0.2, 1.2, -0.2, 1.2]) #x축과 y축 범위를 동시에 지정
```

```python
fig =plt.figure()
ax = fig.add_subplot()
ax.grid(True)
ax.set_xlim([0.5, 2.5])
ax.set_ylim([-10, -5])
```

## 3. Spine - 데이터 영역의 경계를 나타내는 선
- set_position(), set_visible()

```python
x = np.linspace(-10, 10, 100)

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x, x**2)
```

```python
ax.spines['bottom'].set_position(('data',0))
fig
```

```python
ax.spines['left'].set_position(('data',0))
fig
```

```python
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
fig
```

## 4. Axis Scale
- 축의 Scale 을 조정할 수 있다.
- linear
- log
- symlog (= symmetrical log scale)
- logit

```python
x = np.linspace(1, 5, 100)
y = np.power(10, x)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x, y)
```

```python
ax.set_yscale('log')
fig
```

## 5. ShareAxis 기능으로 이중축 그래프를 그릴 수 있다.
- twinx()

```python
np.random.seed(1930)
x = np.arange(10, 110, 10)
y = np.random.randint(1000, 1500, 10)
z = np.random.randint(100, 120, 10)
x,y,z
```

```python
fig = plt.figure()
ax1 = fig.add_subplot() #그래프1
ax1.bar(x, y, label='sales', color='orange', width=5)
ax1.legend(loc='lower center')

ax2 = ax1.twinx() #그래프2, 그래프1과 x축을 Share 한다.
ax2.plot(x,z, label='profit', color = 'magenta')
ax2.legend(loc='lower right')
```
