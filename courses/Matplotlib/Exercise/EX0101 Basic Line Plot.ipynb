## 기본적인 그래프를 그려봅시다.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Data 정의 $ y=2x +1$

```python
x = np.linspace(0, 10, 100) # [0, 10] 사이를 100등분하여 x 를 100개 생성
y = 2*x +1
```

## 2. 그래프 Plotting

```python
fig, ax = plt.subplots() # figure 객체와 axes 객체를 생성

ax.plot(x, y)
```

## 3. linewidth 와 linestyle 을 변경해서 그래프를 그릴 수 있습니다.

```python
fig, ax = plt.subplots() # figure 객체와 axes 객체를 생성

ax.plot(x, y, linewidth=3, linestyle='--')
```

#### 실습1. $y=2x+1$ 의 그래프를 linewidth=5, linestyle='-.' 로 지정해서 그려보세요.

#### 실습2. $y=2x+1$ 의 그래프를 linewidth=1, linestyle='-' 로 지정해서 그려보세요.

## 4. 그래프의 제목과 x축 라벨, y축 라벨을 지정할 수 있습니다.

```python
fig, ax = plt.subplots() # figure 객체와 axes 객체를 생성

ax.plot(x, y)
ax.set_title('y=2x+1')
ax.set_xlabel('x')
ax.set_ylabel('y')
```

#### 실습3. $y=2x+1$ 의 그래프의 제목과 x축 라벨, y축 라벨을 변경해서 그래프를 그려보세요.

#### 실습4. $y=x^2-2x+1$ 의 그래프를 그리고 제목, x축 라벨, y축 라벨을 지정하세요.
