# Ticks (눈금선과 라벨) 에 대해서 알아보자

## 1. Locators
- Locators 는 눈금의 위치를 지정한다.
- 주 눈금(major)와 보조 눈금(minor) 에 대해 각각 설정가능하다.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 100, 1000)
y = x**2
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x, y)
```

```python
print(ax.xaxis.get_major_locator())
print(ax.xaxis.get_minor_locator())
```

### 1. NullLocator : 눈금 없음

```python
from matplotlib.ticker import NullLocator

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x, y)
ax.xaxis.set_major_locator(NullLocator())
```

### 2. MultipleLocator : 특정 값의 배수로 눈금 설정

```python
from matplotlib.ticker import MultipleLocator

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x, y)
ax.xaxis.set_major_locator(MultipleLocator(10))
ax.xaxis.set_minor_locator(MultipleLocator(5))
```

### 3. FixedLocator : 고정값 사용

```python
from matplotlib.ticker import FixedLocator

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x, y)
ax.xaxis.set_major_locator(FixedLocator([0, 20, 30, 40, 60, 100]))
```

## 2. Formatters
- Formatter 는 눈금 라벨을 정의한다. (표시형식)

### 1. NullFormatter : 라벨 없음

```python
from matplotlib.ticker import NullFormatter

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x, y)
ax.xaxis.set_major_formatter(NullFormatter())
```

### 2. PercentFormatter

```python
from matplotlib.ticker import PercentFormatter

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x, y)
ax.xaxis.set_major_formatter(PercentFormatter())
```

### 3. FormatStrFormatter

```python
from matplotlib.ticker import FormatStrFormatter

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x, y)
ax.xaxis.set_major_formatter(FormatStrFormatter("%d yr"))
```
