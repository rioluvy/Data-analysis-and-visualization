# OOStyle code 와 MATLAB Style code 의 차이에 대해 알아보자

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 100) #x 는 0부터 10까지 100등분하여 100개 데이터
```

### 1. Simple Line Plot

#### OOStyle code

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1) # the same as fig.add_subplot(111)

ax1.plot(x, x + 0, linestyle='solid')
ax1.plot(x, x + 1, linestyle='dashed')
ax1.plot(x, x + 2, linestyle='dashdot')
ax1.plot(x, x + 3, linestyle='dotted')
```

#### MATLAB Style code

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');
```

#### OOStyle code

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.plot(x, np.sin(x))
ax1.set_xlabel('x')
ax1.set_ylabel('sin(x)')
ax1.set_title('A Sine Curve')
```

#### MATLAB Style code

```python
plt.plot(x, np.sin(x))
plt.title('A Sine Curve')
plt.xlabel('x')
plt.ylabel('sin(x)')
```

## 2. Scatter Plot

#### OOStyle code

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)

rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    ax1.plot(rng.rand(5), rng.rand(5), marker, \
             label="marker='{0}'".format(marker))

ax1.legend(numpoints=1)
ax1.set_xlim(0, 1.8)
```

#### MATLAB Style code

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
    label="marker='{0}'".format(marker))

plt.legend(numpoints=1)
plt.xlim(0, 1.8)
```

## 3. Multiple Subplots

#### OOStyle code

```python
fig = plt.figure()
fig.subplots_adjust(hspace=0.4, wspace=0.4)
for i in range(1, 7):
    ax = fig.add_subplot(2, 3, i)
    ax.text(0.5, 0.5, str((2, 3, i)),
    fontsize=18, ha='center')
```

#### MATLAB Style code

```python
plt.subplots_adjust(hspace=0.4, wspace=0.4)

for i in range(1, 7):
    plt.subplot(2, 3, i)
    plt.text(0.5, 0.5, str((2, 3, i)),
    fontsize=18, ha='center')
```
