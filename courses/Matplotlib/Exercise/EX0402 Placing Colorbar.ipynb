# 간단한 Colorbar 를 생성해 보자.
- fig.colorbar(ScalarMappable, ax, cax)

```python
import matplotlib.pyplot as plt
import numpy as np; np.random.seed(1)

data = np.random.rand(14,16)
```

## Colorbar 생성하는 2가지 방법
1. 메인 Axes 내부의 영역을 빼앗아 위치 시키기
- fig.colorbar(pad)
- using divider

2. 새로운 Axes 객체를 생성하고 해당 영역에 위치시키기
- subplots() with gridspec_kw
- add_axes()
- grid_spec & add_subplot()

### 1. 메인 Axes 내부의 영역을 빼앗아 위치 시키기

#### fig.colorbar(pad)

```python
fig, ax = plt.subplots()
im = ax.imshow(data)
ax.set_xlabel("x")

cb = fig.colorbar(im, pad=0.05)
```

```python
fig, ax = plt.subplots()
im = ax.imshow(data)
ax.set_xlabel("x")

cb = fig.colorbar(im, orientation='horizontal', pad=0.2)
```

- shrink, aspect 로 길이 줄이기

```python
fig, ax = plt.subplots()
im = ax.imshow(data)
ax.set_xlabel("x")
cb = fig.colorbar(im, orientation='horizontal', pad=0.2, shrink=0.5, aspect=10)
```

#### divider 사용

#### 실습1. 다음 data 에 해당하는 heatmap 을 그리고 colorbar 를 오른쪽에 배치하라.
- data = np.random.rand(8,10)

### 2. 새로운 Axes 객체를 생성하고 해당 영역에 위치시키기

- subplots() with gridspec_kw

```python
fig, (ax, cax) = plt.subplots(2,1, gridspec_kw={"height_ratios":[1, 0.05]})
```

```python
fig, (ax, cax) = plt.subplots(2,1, gridspec_kw={"height_ratios":[1, 0.1]})
im = ax.imshow(data)
ax.set_xlabel("x")

fig.colorbar(im, cax=cax, orientation='horizontal')
```

- add_axes()

```python
fig= plt.figure()
ax = fig.add_axes([0.1,0.35,0.85,0.85])
cax = fig.add_axes([0.2,0.1,0.65,0.10])

im = ax.imshow(data)
ax.set_xlabel("x")
fig.colorbar(im, cax=cax, orientation='horizontal')
```

- grid_spec & add_subplot()

```python
from matplotlib import gridspec as gridspec

fig = plt.figure(figsize=(5,5))
spec = gridspec.GridSpec(25,1, hspace=0.05)

ax = fig.add_subplot(spec[0:18,0])
cax = fig.add_subplot(spec[-2:,0])
```

```python
from matplotlib import gridspec as gridspec

fig = plt.figure(figsize=(5,5))
spec = gridspec.GridSpec(25,1, hspace=0.05)

ax = fig.add_subplot(spec[0:18,0])
cax = fig.add_subplot(spec[-2:,0])

im = ax.imshow(data)
ax.set_xlabel("x")
fig.colorbar(im, cax=cax, orientation='horizontal')
```

#### 실습2. fig.add_axes 로 Axes 객체를 생성한 후, Colorbar 를 해당 위치에 위치시켜라.
- data = np.random.rand(8,10)
- ax = fig.add_axes([0.1,0.35,0.85,0.85])
- cax = fig.add_axes([0.2,0.1,0.65,0.10])
