# 범례상자(Legend) 만드는 방법과 속성들에 대해 알아보자.

## 1. 범례상자를 생성하는 3가지 방법

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,100)
```

### 1) plot 함수 내부에 label 지정

```python
fig = plt.figure()
ax1 = fig.add_subplot()
ax1.plot(x, x+1, label='x+1')
ax1.plot(x, x+2, label='x+2')
ax1.legend()
```

### 2) handle 을 가져와 set_label 사용

```python
fig = plt.figure()
ax1 = fig.add_subplot()
line1, = ax1.plot(x, x+1)
line2, = ax1.plot(x, x+2)
line1.set_label('x+1')
line2.set_label('x+2')
ax1.legend()
```

### 3) handle 을 가져와 Axes.legend 안에서 label 명 부여

```python
fig = plt.figure()
ax1 = fig.add_subplot()
line1, = ax1.plot(x, x+1)
line2, = ax1.plot(x, x+2)
ax1.legend((line1,line2), ('x+1','x+2'))
```

#### 실습1. 다음 A와 B를 표현하는 범례상자를 만들라.
- x = [2018, 2019, 2020, 2021, 2022, 2023]
- A = [115, 120, 128, 130, 140, 152]
- B = [175, 162, 148, 139, 130, 123]

## 2. 범례상자의 속성

```python
x = np.linspace(0, 10, 100)
```

```python
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), '-b', label='Sine')
ax.plot(x, np.cos(x), '--r', label='Cosine')
ax.legend()
```

#### loc : 위치 정보 설정
- 코너 위치 : 'upper left', 'upper right', 'lower left', 'lower right'
- 모서리 위치 : 'upper center', 'lower center', 'center left', 'center right'
- 정중앙 : 'center'

```python
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), '-b', label='Sine')
ax.plot(x, np.cos(x), '--r', label='Cosine')
ax.legend(loc = 'upper right')
```

#### frameon = True/False
- LegendBox 의 외곽선 표시

```python
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), '-b', label='Sine')
ax.plot(x, np.cos(x), '--r', label='Cosine')
ax.legend(loc = 'upper right', frameon=True)
```

#### ncol : 범례의 열 개수

```python
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), '-b', label='Sine')
ax.plot(x, np.cos(x), '--r', label='Cosine')
ax.legend(loc = 'upper right', frameon=True, ncol=2)
```

#### shadow : 그림자 표시

```python
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), '-b', label='Sine')
ax.plot(x, np.cos(x), '--r', label='Cosine')
ax.legend(loc = 'upper right', frameon=True, ncol=2, shadow=True)
```
