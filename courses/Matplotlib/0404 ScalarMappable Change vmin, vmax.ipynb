```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np; np.random.seed(1)
```

### data 값의 범위가 바뀌었을 때, colorbar 도 같이 바뀌어야 할까?

```python
fig = plt.figure(figsize=(10,6))

# HeatMap with data1
data1 = np.random.rand(14,16)
data1_title = 'data1->({:0.3f},{:0.3f})'.format(float(np.min(data1)), float(np.max(data1)))
ax1 = fig.add_subplot(1,2,1)
ax1.set_title(data1_title)
im1 = ax1.imshow(data1)
cb = fig.colorbar(im1, pad=0.05, shrink=0.6)


# HeatMap with data2
data2 = data1*0.4+0.2 #[0.2, 0.6]
data2_title = 'data2->({:0.3f},{:0.3f})'.format(float(np.min(data2)), float(np.max(data2)))
ax2 = fig.add_subplot(1,2,2)
ax2.set_title(data2_title)
im2 = ax2.imshow(data2)
cb = fig.colorbar(im2, pad=0.05, shrink=0.6)
```

### 컬러바(Colorbar) 의 색 범위는 (0,1) 로 고정하고 다시 그려보자.

```python
fig = plt.figure(figsize=(10,6))

# HeatMap with data1
data1 = np.random.rand(14,16)
data1_title = 'data1->({:0.3f},{:0.3f})'.format(float(np.min(data1)), float(np.max(data1)))
ax1 = fig.add_subplot(1,2,1)
ax1.set_title(data1_title)
im1 = ax1.imshow(data1)
# ScalarMappable 객체의 vmin, vmax 를 부여
im1.set_clim(vmin=0., vmax=1.)
cb = fig.colorbar(im1, pad=0.05, shrink=0.6)


# HeatMap with data2
data2 = data1*0.4+0.2 #[0.2, 0.6]
data2_title = 'data2->({:0.3f},{:0.3f})'.format(float(np.min(data2)), float(np.max(data2)))
ax2 = fig.add_subplot(1,2,2)
ax2.set_title(data2_title)
im2 = ax2.imshow(data2)
# ScalarMappable 객체의 vmin, vmax 를 부여
im2.set_clim(vmin=0., vmax=1.)
cb = fig.colorbar(im2, pad=0.05, shrink=0.6)
```

### 좀 더 극명한 대비를 위해 data2->(0.2, 0.6), data3->(0.6, 1.0)

```python
data1 = np.random.rand(14,16)

fig = plt.figure(figsize=(10,6))

# HeatMap with data2
data2 = data1*0.4+0.2 #[0.2, 0.6]
data2_title = 'data2->({:0.3f},{:0.3f})'.format(float(np.min(data2)), float(np.max(data2)))
ax2 = fig.add_subplot(1,2,1)
ax2.set_title(data2_title)
im2 = ax2.imshow(data2)
# ScalarMappable 객체의 vmin, vmax 를 부여
im2.set_clim(vmin=0., vmax=1.)
cb = fig.colorbar(im2, pad=0.05, shrink=0.6)

# change the range of data
data3 = data1*0.4+0.6 #[0.6, 1.0]
data3_title = 'data3->({:0.3f},{:0.3f})'.format(float(np.min(data3)), float(np.max(data3)))
ax3 = fig.add_subplot(1,2,2)
ax3.set_title(data3_title)
im3 = ax3.imshow(data3)
# ScalarMappable 객체의 vmin, vmax 를 부여
im3.set_clim(vmin=0., vmax=1.)
cb = fig.colorbar(im3, pad=0.05, shrink=0.6)
```
