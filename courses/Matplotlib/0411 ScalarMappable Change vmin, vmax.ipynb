```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np; np.random.seed(1)
```

```python
data = np.random.rand(14,16)
print('({:0.3f},{:0.3f})'.format(float(np.min(data)), float(np.max(data))))
```

```python
fig, ax = plt.subplots()
ax.set_xlabel("x")

im = ax.imshow(data)
cb = fig.colorbar(im, pad=0.05)
```

#### If we change the color limits by using set_clim with ScalarMappable, the colors of contents will be changed.

```python
fig, ax = plt.subplots()
ax.set_xlabel("x")
im = ax.imshow(data, vmin=0.2, vmax=0.6)
#im.set_clim(vmin=0.2, vmax=0.6)
cb = fig.colorbar(im, pad=0.05)
```

### How to change the colorbar limits keeping the color of contents unchanged ?

```python
data = np.random.rand(14,16)

# change the range of data
data = data*0.4+0.2 #[0.2, 0.6]
print('({:0.3f},{:0.3f})'.format(float(np.min(data)), float(np.max(data))))
```

```python
fig, ax = plt.subplots()
ax.set_xlabel("x")
im = ax.imshow(data)
cb = fig.colorbar(im, pad=0.05)
```

### Problem: the color of contents is normalized in (0.2, 0.6) & colorbar follows the range

### Fix the ScalarMappable color limits (0, 1.)

```python
fig, ax = plt.subplots()
ax.set_xlabel("x")
im = ax.imshow(data, vmin=0., vmax=1.)
im.set_clim(vmin=0., vmax=1.)
cb = fig.colorbar(im, pad=0.05)
```

```python
# assign norm and cmap
fig, ax = plt.subplots()
ax.set_xlabel("x")
cmap = mpl.cm.get_cmap('viridis')
norm = mpl.colors.Normalize(vmin=0., vmax=1.0)
#vmin, vmax : float or None
#    If *vmin* and/or *vmax* is not given, they are initialized from the
#    minimum and maximum value, respectively

im = ax.imshow(data, norm=norm, cmap=cmap)
cb = fig.colorbar(im, pad=0.05)
```

```python
print(cmap(0.))
print(cmap(0.5))
print(cmap(1.))
```

### 3 Ways to draw a Colorbar
- 1) ScalarMappable
- 2) ColorbarBase
- 3) ScalarMappable "on the fly"

```python
data = np.random.rand(14,16)

# change the range of data
data = data*0.4+0.2 #[0.2, 0.6]
np.min(data), np.max(data)
```

```python
# ScalarMappable
fig, ax = plt.subplots()
ax.set_xlabel("x")
im = ax.imshow(data, vmin=0., vmax=1.)
cb = fig.colorbar(im, pad=0.05)
```

```python
#ColorbarBase
fig = plt.figure()
ax = fig.add_axes([0.04, 0.04, 0.82, 0.8])
cax = fig.add_axes([0.8, 0.04, 0.06, 0.75])
ax.set_xlabel("x")
im = ax.imshow(data, vmin=0. , vmax=1.)

cmap = mpl.cm.get_cmap('viridis')
norm = mpl.colors.Normalize(vmin=0., vmax=1.0)

mpl.colorbar.ColorbarBase(cax, cmap=cmap, norm=norm)
```

```python
# ScalarMappable "on the fly"
fig = plt.figure()
ax = fig.add_axes([0.04, 0.04, 0.82, 0.8])
cax = fig.add_axes([0.8, 0.04, 0.06, 0.75])
ax.set_xlabel("x")
im = ax.imshow(data, vmin=0. , vmax=1.)

cmap = mpl.cm.get_cmap('viridis')
norm = mpl.colors.Normalize(vmin=0., vmax=1.0)
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)

fig.colorbar(sm, cax=cax)
```
