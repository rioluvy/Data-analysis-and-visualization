```python
x = np.linspace(0,10,100)
y = x **2
```

```python
fig, ax1 = plt.subplots()
ax1.plot(x,y)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.plot(x,y)
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax1.plot(x,y)

ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
ax2.plot(x, np.sqrt(x))
```
