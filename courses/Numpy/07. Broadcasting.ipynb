# 7. Broadcasting
- Numpy Broadcasting 에 대해 알아보자.

1) 일차원 배열 + 스칼라  
2) 2차원 배열 + 일차원 배열  
3) 브로드캐스팅 룰

## 1. 일차원 배열 + 스칼라
- 일차원 배열에 스칼라 연산이 어떻게 이뤄지는지 확인해 보자.

```python
a = np.arange(4)
print(a.shape)
```

```python
s = 10
```

```python
b = a + s
```

```python
print(b.shape)
```

## 2. 2차원 배열 + 일차원 배열

```python
a = np.arange(8).reshape(2,4)
b = np.arange(4)
print(f'a.shape:{a.shape}, b.shape:{b.shape}')
```

```python
c = a+b
c
```

```python
print(f'c.shape:{c.shape}')
```

## 3. 브로드캐스팅 룰
- Rule1. 차원이 모자라는 경우 1로 추가(pad) 한다.
- Rule2. 차원의 원소의 개수가 다른 경우 1을 큰 수에 맞춘다.
- Rule3. 차원의 원소의 개수가 다른데 1이 없는 경우 Error

```python
a = np.arange(24).reshape(2,3,4)
b = np.ones((3,1))
c = np.zeros((1,4))
```

```python
d = a+b+c
d
```

```python
print(f'd.shape:{d.shape}')
```

**Error Case**

```python
a = np.arange(24).reshape(2,3,4)
b = np.ones((3,1))
c = np.zeros((1,2))
```

```python
d = a+b+c
```
