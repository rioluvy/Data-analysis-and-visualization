# 4. Array arithmetic and element-wise operations
- array 연산에 대해 확인해 보자.

1) 기본적인 사칙연산
2) 삼각함수
3) 로그함수와 지수함수
4) Rounding

## 1. 기본적인 사칙연산

```python
a = np.arange(1, 10).reshape((3,3))
b = np.arange(10,19).reshape((3,3))

addition = a + b
subtraction = a - b
multiplication = a * b
true_division = a / b
floor_division = a // b
```

```python
print(f'Array a:\n{a}\n')
print(f'Array b:\n{b}\n')
print(f'a + b:\n{addition}\n')
print(f'a - b:\n{subtraction}\n')
print(f'a * b:\n{multiplication}\n')
print(f'a / b:\n{true_division}\n')
print(f'a // b:\n{floor_division}\n')
```

## 2.삼각함수

```python
x = np.linspace(-4, 4, 200)

# sin function
sin = np.sin(x)
# cosine function
cos = np.cos(x)
# tangent function
tan = np.tan(x)
```

```python
import matplotlib.pylab as plt
plt.style.use('dark_background')
%config InlineBackend.figure_format = 'retina' # to get high resolution images
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, 1)
ax1.plot(x, sin)
ax1.set_title("sin")
ax2.plot(x, cos)
ax2.set_title("cos")
ax3.plot(x, tan)
ax3.set_title("tan")
plt.tight_layout()
```

## 3. 로그함수와 지수함수

```python
z = np.array([0.1, 1, np.e, np.pi])

# exponent
exp = np.exp(z)
# exponent(x) -1
expm1 = np.expm1(z)
# 2^P
exp2 = np.exp2(z)
# natural log
log = np.log(z)
# log base 10
log10 = np.log10(z)
# log base 2
log2 = np.log2(z)
```

```python
print(f'exp(x):\n{exp}\n')
print(f"exp(x) - 1:\n{expm1}\n")
print(f"2**x:\n{exp2}\n")
print(f'log(x):\n{log}\n')
print(f'log_10(x):\n{log10}\n')
print(f'log_2(x):\n{log2}\n')
```

## 4. Rounding

```python
decimals = np.linspace(0.11111111, 0.99999999, 10)

# rounding
around = np.around(decimals, 3)
# rounding
round_ = np.round(decimals, 3)
# rounding to integer
rint = np.rint(decimals)
# round to the floor
floor = np.floor(decimals)
# round to the ceiling
ceil = np.ceil(decimals)
```

```python
print(f"decimals:\n{decimals}\n")
print(f"'around' round to the given number of decimals:\n{around}\n")
print(f"'round' yields identical results than 'around':\n{round_}\n")
print(f"'rint' round to the nearest integer:\n{rint}\n")
print(f"'floor' round to the floor of the input:\n{floor}\n")
print(f"'ceil' round to the ceiling of the input:\n{ceil}")
```
