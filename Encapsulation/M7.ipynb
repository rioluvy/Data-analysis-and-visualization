```python
import seaborn as sns
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
```

```python
## Pandas 기본 Plot()함수
# 수치형 데이터프레임에서 plot()함수를 실행
# df.plot()함수는 index를 기준으로 line graph를 출력
# Arguments of Plot()
# kind: 그래프의 타입을 결정: 'bar', 'box' 등..
# figsize: output 그림의 크기
# fontsize: chat내 글자 크기

import matplotlib.pyplot as plt

# Line plot
df1.plot()
plt.title("Pandas의 plot메소드 사용 예")
plt.xlabel("시간")
plt.ylabel("Data")
plt.show()
```

```python
# Bar chart: line chart와 다르게 각 index당 정확한 수치를 보여줌

iris.sepal_length[:20].plot(kind = 'bar', rot=0)
plt.title("Visualization of sepal_length")
plt.xlabel("Data")
plt.ylabel("sepal_length")
plt.show()
```

```python
iris[:5].plot.barh(rot=0) ## barh: 수평 막대 그래프
plt.title("Iris 데이터의 bar plot")
plt.xlabel("Data")
plt.ylabel("각 Feature의 값")
plt.show()
```

```python
# Bar chart with groupby()
df2 = iris.groupby(iris.species).mean()
df2.columns.name = "feature"

df2.plot.bar(rot=0)
plt.title("각 종의 Feature별 평균")
plt.xlabel("mean")
plt.ylabel("kinds")
plt.ylim(0,8)
plt.show()
```

```python
# pie chart
# kind = 'pie'
# 특정 컬럼 내에서 category가 차지하는 비율을 시각화
# 그룹간의 차이가 명확할 수록 가시성이 좋음
# 그룹의 수가 과도하게 많지 않은 컬럼에 대해 적용

df3 = titanic.pclass.value_counts()
df3.plot(kind='pie',autopct = "%.2f%%")
plt.title("선실별 승객 수 비율")
plt.axis('equal')
plt.show()
```

```python
# histogram
# kind = 'hist' or plot.hist()
# 나이를 볼때 많이 사용함.
# 전체적인 분포를 시각화
# 연속성 데이터를 사용자가 지정한 범위로 압축
# 그래프의 개형을 통해 어느 구간에 표본이 집중되는지 보기 좋음

iris.plot.hist()
plt.title("각 Feature 값들의 빈도수 histogram")
plt.xlabel("데이터 값")
plt.show()
```

```python
# 커널 밀도 함수(KDE)
# kind = 'kde' or plot.kde()
# 히스토그램이 박스 단위로 분포를 보였다면, 연속적인 분포를 확인
# 분포에 근사한 커널 밀도 함수의 개형을 제공

iris.plot.kde()
plt.title("각 Feature 값들의 빈도수에 대한 Kernal Density Function")
plt.xlabel("데이터 값")
plt.show()
```

```python
# box chart
# kind = 'box' or plot.box()
# 각 컬럼별 표본의 분포를 박스의 형태로 나타냄
# 각 박스는 1~4분위 값, 평균값, 최대, 최소 값을 한번에 전달
# 다양한 분포 정보를 쉽게 전달함

iris.plot.box()
plt.title("각 Feature 값들의 빈도수에 대한 Box plot")
plt.xlabel("Feature")
plt.ylabel("데이터 값")
plt.show()
```

```python
# Titanic 데이터에서 확인했던 차이들을 pandas plot으로 표현해보자
# pclass와 사용한 Fare을 여러 차트로 표현해보자
# Sex 와 Survived의 counts를 표현해보자
# Pie chart로 각 성별 생존 비율을 그려보자
# 나이를 구간이 아닌 범주형으로 바꾸어 특징을 그려보자

```

```python
# scatter plot (산점도)
# kind = 'scatter'
# 표본의 각 하나의 데이터를 점에 대응시켜 좌표에 나타냄
# 각 점들의 분포를 여러 축으로 확장하여 관찰 가능
# 단 너무 많은 데이터를 그릴 시 가독성이 떨어질 수 있음
# 2차원 차트이기 때문에 x와 y값을 입력받음

iris.plot.scatter(x = 'sepal_length', y = 'sepal_width')
plt.title('각 Feature의 종 별 데이터에 대한 Scatter plot')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Titanic 데이터 불러오기 (Seaborn 사용)
titanic_data = sns.load_dataset('titanic')

# 그래프 레이아웃 설정
plt.figure(figsize=(12, 8))

# 성별에 따른 생존 여부
ax1 = plt.subplot(221)
sex_survived = titanic_data[['sex', 'survived']]
sex_survived.plot(kind='bar', x='sex', y='survived', ax=ax1)
ax1.set_title('성별에 따른 생존 여부')

# 클래스별 나이 분포
ax2 = plt.subplot(222)
class_age = titanic_data.groupby('class')['age']
class_age.hist(ax=ax2)
ax2.set_title('클래스별 나이 분포')

# 요금과 나이 사이의 관계
ax3 = plt.subplot(223)
fare_age = titanic_data[['fare', 'age']]
fare_age.plot(kind='scatter', x='fare', y='age', ax=ax3)
ax3.set_title('요금과 나이 사이의 관계')

# 그래프 출력
plt.tight_layout()
plt.show()
```

```python
ax2 = plt.subplot(222)
class_age = titanic_data[['class','age']]
class_age.plot(kind = 'hist', ax = ax2)
ax2.set_title("Age distribution by class")
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Titanic 데이터 불러오기 (Seaborn 사용)
titanic_data = sns.load_dataset('titanic')

plt.figure(figsize = (14,14)) ## 가로 세로 길이 설정
# 1. 성별(sex)에 따른 생존 여부(survived)를 막대 그래프로 시각화 (Pandas의 plot 함수 사용)
ax1 = plt.subplot(221) ## 2행 2열 1번째 서브플롯
# 2 : 전체 서브플롯의 행 수(2행)
# 2: 전체 서브플롯의 열 수(2열)
# 1: 현재 활성화할 서브플롯의 위치(1번째 서브플롯)
sex_survived = titanic_data[['sex','survived']]
sex_survived.plot(kind = 'bar', x = 'sex', y = 'survived', ax = ax1)
ax1.set_title("survival by sex")

# 2. 클래스(class)에 따른 나이(age) 분포를 히스토그램으로 시각화 (Pandas의 plot 함수 사용)
ax2 = plt.subplot(222)
class_age = titanic_data.groupby('class')['age']
class_age.plot(kind = 'hist', ax = ax2)
ax2.set_title("Age distribution by class")

# 3. 요금(fare)과 나이(age) 사이의 관계를 산점도로 시각화 (Pandas의 plot 함수 사용)
ax3 = plt.subplot(223)
fare_age = titanic_data[['fare','age']]
fare_age.plot(kind = 'scatter', x = 'fare', y = 'age', ax = ax3)
ax3.set_title("relationship between fare and age")
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Iris 데이터 불러오기 (Seaborn 사용)
iris_data = sns.load_dataset('iris')

# Iris 데이터를 Pandas DataFrame으로 변환
iris_df = pd.DataFrame(iris_data)

# 그래프 레이아웃 설정
plt.figure(figsize=(12, 8))

# 1. 꽃 종(species)에 따른 꽃받침 길이(sepal_length)와 꽃받침 너비(sepal_width)의 관계를 산점도로 시각화
ax1 = plt.subplot(221)
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris_df)
ax1.set_title('꽃 종에 따른 꽃받침 길이와 너비의 관계')

# 2. 꽃 종(species)에 따른 꽃잎 길이(petal_length)와 꽃잎 너비(petal_width)의 관계를 산점도로 시각화
ax2 = plt.subplot(222)
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=iris_df)
ax2.set_title('꽃 종에 따른 꽃잎 길이와 너비의 관계')

# 3. 꽃 종(species)에 따른 꽃받침 길이(sepal_length)의 분포를 히스토그램으로 시각화
ax3 = plt.subplot(223)
sns.histplot(x='sepal_length', hue='species', data=iris_df)
ax3.set_title('꽃 종에 따른 꽃받침 길이의 분포')

# 4. 꽃 종(species)에 따른 꽃잎 길이(petal_length)의 분포를 히스토그램으로 시각화
ax4 = plt.subplot(224)
sns.histplot(x='petal_length', hue='species', data=iris_df)
ax4.set_title('꽃 종에 따른 꽃잎 길이의 분포')

# 그래프 출력
plt.tight_layout()
plt.savefig('iris_visualization.png')
```

```python
import seaborn as sns
import pandas as pd

# Iris 데이터 불러오기 (Seaborn 사용)
iris_data = sns.load_dataset('iris')

# Iris 데이터를 Pandas DataFrame으로 변환
iris_df = pd.DataFrame(iris_data)

plt.figure(figsize = (12,12))
# 1. 꽃 종(species)에 따른 꽃받침 길이(sepal_length)와 꽃받침 너비(sepal_width)의 관계를 산점도로 시각화
ax1 = plt.subplot(221)
sns.scatterplot(x = 'sepal_length', y = 'sepal_width', hue = 'species', data = iris_df)
ax1.set_title('The relationship between the length and width of the flower stand according to the flower species')

# 2. 꽃 종(species)에 따른 꽃잎 길이(petal_length)와 꽃잎 너비(petal_width)의 관계를 산점도로 시각화
ax2 = plt.subplot(222)
sns.scatterplot(x = 'petal_length', y = 'petal_width', hue = 'species', data = iris_df)
ax2.set_title('The relationship between petal length and width according to flower species')

# 3. 꽃 종(species)에 따른 꽃받침 길이(sepal_length)의 분포를 히스토그램으로 시각화
ax3 = plt.subplot(223)
sns.histplot(x = 'sepal_length', hue = 'species', data = iris_df)
ax3.set_title('꽃 종에 따른 꽃받침 길이의 분포')

# 4. 꽃 종(species)에 따른 꽃잎 길이(petal_length)의 분포를 히스토그램으로 시각화
ax4 = plt.subplot(224)
sns.histplot(x = 'petal_length', hue = 'species', data = iris_df)
ax4.set_title('꽃 종에 따른 꽃잎 길이의 분포')

# 그래프를 이미지 파일로 저장하세요.
plt.tight_layout()  ## 가로세로 축을 자동으로 조정하여 깔끔하게 보이도록 해줌
plt.savefig('iris_visualization.png')
```

```python
import pandas as pd
import matplotlib.pyplot as plt


def main():
    # Step 1: 데이터셋 생성
    data = {
        'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
        'Sales': [200, 180, 220, 240, 210, 230, 300, 320, 280, 260, 180, 150],
        'Expenses': [150, 140, 170, 160, 150, 180, 190, 200, 210, 180, 160, 140]
    }
    df = pd.DataFrame(data)


    # Step 2: 선 그래프로 판매량과 지출 시각화
    sales_expense = df[['Sales', 'Expenses']]
    sales_expense.plot()
    plt.title('Monthly Sales and Expenses')
    plt.ylabel('Amount')
    download_figure('SalesNExpenses')
    plt.show()

    # Step 3: 원형 그래프로 각 달의 판매량 비율 시각화
    df.set_index('Month', inplace = True)
    plt.pie(df['Sales'], labels=df.index, autopct='%.1f%%')
    plt.title('Sales Distribution by Month')
    plt.ylabel('')
    download_figure('SalesDist')

    plt.show()

def download_figure(filename):
    plt.savefig(f"{filename}.png")


if __name__ == "__main__":
    main()
```
