```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## Seaborn으로 데이터 시각화 하기
# matplotlib을 더 쉽게 다루기 위한 high level의 library
# 장점: 간결함
# matplotlib이 더 정교하게 컨트롤 할 수 있지만, seaborn이 편하고 가시성이 좋다
```

```python
## Seaborn
# Seaborn은 Matplotlib을 사용하는 High-level 라이브러리
# 여러 통계 정보와 데이터 분석에 사용됨
# matplotlib과 비교하여 간결한 사용성
# 그래프 스타일 변경의 편리성
# matplotlib에서 제공하는 거의 대부분의 함수를 사용 가능
# load_dataset()을 통한 샘플 데이터 제공
```

```python
## 테마 선정
# 차트의 배경 테마를 선택 가능
# Default: 'darkgrid'
# whitegrid, white, ticks 등 선택 가능
# seaborn의 set()함수 사용 + style = 테마명
# sns.set(style = whitegrid)
```

```python
## Line plot
# 기존 amtplotlib의 plt.plot()과 동일한 작동
# sns.lineplot() 함수 사용
# matplotlib과 동일하게 여러개의 선 사용 가능
# 다변량 그래프의 경우 데이터를 직접 함수에 넣어서 사용

iris = sns.load_dataset('iris')
x = iris.petal_length.values
sns.lineplot(x)

# sns.lineplot(data = iris, x = 'sepal_length', y = 'sepal_width')
# plt.title('Iris Rug plot')
plt.show()
```

```python
## Single Variate Plot
# 하나의 차원의 데이터의 분포를 나타내는 차트
# 실수 분포 플롯은 자료의 분포를 묘사
# Matplotlib의 단순한 히스토그램과 달리 커널 밀도 및 러그 표시 기능 및 다차원 복합 분포 기능 등을 제공
# rugplot(), kdeplot(), distplot()

## Rugplot()
# 데이터의 위치를 x축 위에 작은 선분(rug)로 표현
# Scatter plot등 다른 plot함수와 함께 사용 가능

sns.scatterplot(x = 'petal_length', y = 'petal_width', data = iris)
sns.rugplot(x = 'petal_length', y = 'petal_width', data = iris)
plt.show()

## Violin plot
# 데이터셋의 카테고리별 분포를 시각화
# matplotlib의 바이올린 플롯과 동일
# 중심에 대칭인 kdeplot
# 통계수치(흰색점: 중앙값, 굵은 중앙선: 사분위선, 얇은 중앙선: 신뢰구간)
sns.violinplot(x = 'petal_length', data = iris)
plt.show()

## KDE plot
# 커널 밀도 함수를 사용하여 데이터셋의 분포 시각화
# 히스토그램과 비교했을 때 연속적인 분포곡선
# 1,2차원의 데이터 모두 표현 가능
# sns.kdeplot(data=iris, x = 'petal_width') 1차원
sns.kdeplot(data = iris, x = 'petal_width', y = 'petal_length') ## 2차원
plt.show()
```

```python
## Catplot
# 카테고리 데이터의 분포를 표현
# x: 사용할 카테고리 컬럼
# y: 수치형 데이터
# col: 카테고리 세분화
# kind: 분포도 선택(box)
tips = sns.load_dataset('tips')
sns.catplot(data = tips, x = 'day', y = 'total_bill', col = 'smoker') ## smoker T,F에 따른 플롯
```

```python
## Countplot
# 카테고리 별 갯수를 카운팅하여 시각화, 갯수에 초점
# countplot() 함수는 데이터프레임에만 사용가능
# countplot(x = 'column_name, data = dataframe)
# hue: 값에 다른 컬럼을 추가하여 변수 추가
titanic = sns.load_dataset('titanic')
sns.countplot(x = 'class', data = titanic, hue = 'sex')
plt.title("타이타닉 호의 각 클래스 별 승객 수")
plt.show()
```

```python
## heatmap
# 하나의 변량이 아닌 여러개의 인자의 분포에 대한 정보 전달
# 2개 이상의 변량에 대해, 각 경우에 따른 점수를 색상으로 표현
# 피봇테이블의 경우 인자로 데이터프레임 사용
# sns.heatmap()사용
# annot: True설정시 수치 작성 가능

data = pd.DataFrame(data = {
    'A': [1,4,3,6,5,8,7,9],
    'B': [6,5,7,8,9,9,8,9],
    'C': [8.8,7.7,6.6,5.5,4.4,3.3,2.2,1.1]
})
sns.heatmap(data, annot = True)
```

```python
## jointplot()
# Scatterplot으로 전달하고자 하는 정보가 부족할 경우
# jointplot은 jointplot명령은 scatterplot뿐 아니라 차트의 가장자리(margin)에 각 변수의 히스토그램도 그린다
# kind값을 변경하여 그래프 형식 변경
sns.jointplot(x = 'sepal_length',y = 'sepal_width',data=iris)
plt.suptitle("IRIS Joint Plot", y= 1)
plt.show()
```

```python
## Pairplot
# 3차원 이상의 데이터라면 pairplot명령을 사용
# pairplot은 그리드 형태로 각 데이터 열의 조합에 대해 플롯을 그려줌(3차원 데이터라면 2개 2개씩 짝 지은 쌍 그래프를 모두 표현)
# 대각선 영역에는 해당 데이터의 히스토그램을 그린다
sns.pairplot(iris[['petal_width','sepal_length','sepal_width']])
plt.title("iris Data's pair plot")
plt.show()
```

```python
## Relational plots: relplot()
# scatterplot과 lineplot모두 사용가능
# kind값을 조정하여 원하는 스타일 설정(line, scatter)
# relplot()의 리턴값은 AxesSubplot이 아닌 FacetGrid(일반 line과 scatter와의 차이점)
# 그 외 사용법은 scatter, line과 동일
sns.relplot(data=titanic, x = 'age', y='fare',hue='pclass')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Iris 데이터 불러오기
iris_data = sns.load_dataset('iris')

# Heatmap 그리기
correlation_matrix = iris_data.drop(columns = ['species']).corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')

# Violinplot 그리기
sns.violinplot(x='species', y='petal_length', data=iris_data)

# 그래프 출력
plt.title('Heatmap & Violinplot (Iris Data)')
plt.show()
```
