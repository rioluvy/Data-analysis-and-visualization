```python
# 1. how many nulls
print(titanic.isnull().sum())
```

```python
titanic['age'].fillna(titanic['age'].median(), inplace = True)
titanic['embarked'].fillna(titanic['embarked'].mode()[0], inplace = True)
```

```python
print(titanic.isnull().sum())
```

```python
titanic['age'] = titanic['age'].astype(int)
```

```python
titanic.head()
```

```python
titanic['familySize'] = titanic['sibsp'] + titanic['parch'] + 1
```

```python
titanic.head()
```

```python
# 그룹으로 묶어보자
grouper = titanic.groupby(['class'])
grouper['survived'].sum().reset_index()
```

```python
grouper['survived'].mean().reset_index()
# 일등석은 62퍼로 생존
# 이등석은 47퍼
# 삼등석은 24퍼
```

```python
titanic.groupby(['embark_town'])['fare'].mean().reset_index()
# Cherbourg에 정박한 사람들이 많은 비용을 내고 있음
```

```python
titanic.groupby(['embark_town']).agg({
    'fare':'mean',
    'age':['median','min']
}).reset_index()
```
