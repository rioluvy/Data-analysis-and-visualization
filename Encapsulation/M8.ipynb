```python
## Pandas에서는 plot()함수를 통해 간단하게 시각화를 해봤음
## Matplotlib vs Pandas.plot()
# pandas.plot()은 사용이 간편하다. but 세부적인 조정이 어려움
# Matplotlib을 활용하면 pandas plot() 차트를 세부적으로 조정 가능
```

```python
## Matplotlib
# 데이터 시각화와 2D 그래프를 생성을 위한 파이썬 라이브러리
# Pandas 데이터 프레임을 활용하여 여러가지 차트를 쉽게 생성 가능
# 여러가지 형식의 차트를 지원하며 자유롭게 생성 가능

## Matplotlib에서 생성 가능한 plot
# 기본차트: line, scatter, bar
# 통계차트: histogram, box, errorbar
# 그 외: 3D, 좌표 데이터, 배열과 벡터 등

## plt.plot(): default는 line
## plt.sactter(): 산점도
## 그래프 함수 호출 이후, 최종적으로 차트를 출력하기 위해 plt.show() 사용

## 차트를 figure라는 공간으로 관리
## fig = plt.figure()
## 최종 figure는 savefig()함수를 통해 파일로 저장 가능(fig.savefig('figure.png'))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
## Line plot
# 함수 호출: plt.plot([1,2,3,4])
# 차트 출력: plt.show()
plt.title('line plot')  ## 차트의 제목 설정
plt.plot([1,3,5,7,9]) ## python array뿐만 아니라
plt.show()

## plot()함수 내부에는 x값 혹은 x,y값을 넣어줄 수 있다
# 하나의 배열만 입력: x는 1~N까지 자동으로 지정해서 plot생성
# 2개의 동일한 길이의 배열 입력: 각 x값에 따른 y값들을 선으로 잇는 plot생성
plt.title('line plot')
plt.plot([1,3,5,7,9],[1,2,10,5,29]) ## (1,1), (3,2), (5,10), (7,5), (9,29)
plt.show()

x = np.linspace(0,10,100)
fig = plt.figure()
plt.plot(x,np.sin(x),linestyle = '-', color='blue', marker='.') ## x,y,chat 모양, 색깔, data point의 표시(점마커)
plt.plot(x,np.cos(x),linestyle = '--', color='magenta', marker='o') ## 원형 마커(v,x 등 여러가지 모양이 있음)
plt.show()
```

```python
## scatter plot: 산점도: 데이터 포인트를 하나의 점으로 나타내는 plot
# plt.scatter()함수를 호출하여 사용
# 입력으로는 x,y값을 받아오며, x,y는 동일한 길이의 배열
# 각 x와 y값에 대응되는 점을 plot한다

rng = np.random.RandomState(0)
x = np.linspace(0,10,30)  ## arr
y = np.sin(x) ## arr
colors = rng.randn(30)
size = rng.randn(30)*50
plt.scatter(x,y, s = size, c = colors) ## x,y,size(반지름의 크기), color(size와 color도 arr로 넣으면 점들의 색상도 구분 가능)
```

```python
## Bar plot: 데이터 포인트의 수치를 막대로 표현
# plt.bar()함수를 호출하여 사용
# 입력으로는 x, height값을 받아오며, x,height는 동일한 길이의 배열
# 각 x값에 대응되는 높이의 막대를 표현
# 추후 배울 tick옵션을 활용해 문자열을 x축으로 사용 가능

y = [2,3,1]
x = np.arange(len(y))
plt.title("bar chart")
plt.bar(x,y)
plt.show()

x = [1,2,3,4]
y = rng.randn(4)
w = rng.randn(4)
plt.bar(x,y,width=w,color='orange') ## width:막대의 너비, botto,: 막대의 시작점을 지정, color
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Titanic 데이터 불러오기 (Seaborn 사용)
titanic_data = sns.load_dataset('titanic')

# 1. 나이(Age)에 따른 승객 수를 선 그래프로 시각화
tmp = titanic_data['age'].value_counts().sort_index()
x = tmp.index
y = tmp.values
plt.plot(x,y)
plt.show()

# 2. 성별(Sex)에 따른 생존 여부(Survived)를 선 그래프로 시각화
tmp = titanic_data.groupby('sex')['survived'].mean().reset_index()
x = tmp['sex']
y = tmp['survived']
plt.plot(x,y, marker = "o")
plt.show()


# 3. 객실 등급(Pclass)에 따른 승객 수를 막대 그래프로 시각화
tmp = titanic_data['pclass'].value_counts().sort_index()
x = tmp.index
y = tmp.values
plt.bar(x,y)
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def plot_image():
    data = {
        'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'],
        'Math': [85, 90, 78, 92, 88],
        'English': [92, 88, 95, 80, 78],
    }

    df = pd.DataFrame(data)

    # 1. Scatter plot
    plt.figure(figsize = (8,6))
    x = df['Math']
    y = df['English']
    plt.scatter(x,y)
    plt.title('Student Math and English Performance Scatter Plot')

    # 테스트 코드 (수정 금지)
    plt.savefig("scatter.png")

    # 2. Bar chart
    plt.figure(figsize=(8,6))
    x = df['Name']
    y = df['Math']
    plt.bar(x,y)
    plt.title('Student Math Performance Bar Chart')

    # 테스트 코드 (수정 금지)
    plt.savefig("bar.png")


plot_image()



# figure 사이즈는 (8, 6)입니다.
# edgecolor는 검정 (‘k’)입니다.
# color는 파랑 (‘b’)입니다.
# Scatter Plot
# x축을 수학 성적으로, y축을 영어 성적으로 plot 해주세요.
# Plot title은 ‘Student Math and English Performance Scatter Plot’ 입니다.
# Bar Chart
# 수학 성적에 대한 bar chart를 그려주세요.
# Plot title은 ‘Student Math Performance Bar Chart’입니다.
```
